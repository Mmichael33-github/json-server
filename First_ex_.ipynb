```python

import pandas as pd
import numpy as np
from google_play_scraper import app, Sort, reviews_all
import matplotlib.pyplot as plt
import json
```

```python
hk_prjct = reviews_all('com.hikingproject.android', sleep_milliseconds=0, lang = 'en', country = 'US', sort=Sort.NEWEST)
```

```python
df = pd.DataFrame(hk_prjct)
df
```
